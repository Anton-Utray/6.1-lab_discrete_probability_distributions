## Bernoulli's Distribution

In a basket containing 100 fruits, there are 60 apples and 40 oranges. 

A fruit is picked randomly from the basket. 

1. **What is the probability that the fruit is apple?**
1. **What is the probability that it is an orange?**

````python
from scipy.stats import bernoulli
````

````python
"""
Calculate:
p = probability that the fruit is an apple = Number of apples / Total number of fruits
q = probability that the fruit is an orange = Number of oranges / Total number of fruits
"""

p=60/100
q=40/100
````

````python
jacob_b = bernoulli(p)
muestra = jacob_b.rvs(size=100)
muestra
````

````python
import seaborn as sns

sns.countplot(x=muestra);
````

**Plot the PDF (probability distributiom function) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
# your code here
# Please label the axes and give a title to the plot 

x=[i for i in range(5)]                  # nº de exitos

y=[binom.cdf(x, 20, p) for x in x]      # su probabilidad

plt.plot(x, y)


plt.xlabel('nº of apples being picked')
plt.ylabel('probability')
plt.title('probability of less than 5 fruits picked are apples');
````
